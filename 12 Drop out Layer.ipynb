```python
#in this we don't want any header
df=pd.read_csv("C:\\Users\\user\\Downloads\\sonar_csv.csv",header=None)
df
```

```python
X=df.drop(60,axis="columns")
X
```

```python
y=df[60] #target
y
```

```python
#we have to convert this things to numbers,since they are in strings
```

```python
y=pd.get_dummies(y,drop_first=True)
y
```

```python
y.value_counts()  #mine=111,rock=97
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=8,epochs=100) #batch size is first they will take 8 batch,then train,then again batch of 8
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test).reshape(-1)
```

```python
#Adding dropout layer
```

```python
model2=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model2.fit(X_train,y_train,batch_size=8,epochs=100) #batch size is first they will take 8 batch,then train,then again batch of 8
```

```python
model2.evaluate(X_test,y_test)
```
