```python
#function to remove punctuatuions
import string
from string import punctuation
def remove_punc(text):
    for i in text:
        if i in punctuation:
            text=text.replace(i,"")
    return text     
    
```

```python
train["text"]=train.text.map(lambda x:remove_punc(x))
```

```python
train["text"]
```

```python
#function to remove URLs:
import re
def remove_url(text):
    return re.sub(r"http\S+","",text) #\S means all letters until a space comes,+ means more than 1
```

```python
train["text"]=train.text.map(lambda x:remove_url(x)) #map is used instead of a loop
```

```python
#removing emojis
import emoji
```

```python
emoji.UNICODE_EMOJI #ALl the emoji list
```

```python
def remove_emoji(text):
    allchars=[k for k in text]
    emoji_list=[c for c in allchars if c in emoji.UNICODE_EMOJI]
    clean_text=" ".join([j for j in text.split() if not any (i for i in emoji_list)])
    return clean_text
    #any return true if the item is there,else it will give false
```

```python
train["text"]=train.text.map(lambda x:remove_emoji(x))
train["text"]
```

```python
#removing stopwords
from nltk.corpus import stopwords
stop=set(stopwords.words("english"))#set will create unique stopwords
```

```python
def remove_stopwords(text):
    [i.lower() for i in text.split() if i.lower() not in stop]
    return "".join(text)
```

```python
train["text"]=train.text.map(remove_stopwords)
train["text"]
```

```python
from collections import Counter
```

```python
train["text"].values
```

```python
def counter_word(text):
    counter=Counter()
    for i in text.values:
        for word in i.split():
            counter[word]+=1
    return counter        
        
```

```python
text=train.text
```

```python
counter=counter_word(text)
counter
```

```python
len(counter)
```

```python
num_words=len(counter)
```

```python
max_length=20 #number of words in a sequence
```

```python
train_size=int(train.shape[0]*0.8) #training size
```

```python
train_sentences=train.text[:train_size] #training text
train_labels=train.target[:train_size] #training targets
```

```python
test_sentences=train.text[train_size:]
test_labels=train.target[train_size:]
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer(num_words=num_words)
```

```python
tokenizer.fit_on_texts(train_sentences)
```

```python
train_sequences=tokenizer.texts_to_sequences(train_sentences)
train_sequences 
```

```python
#padding
```

```python
from keras.preprocessing.sequence import pad_sequences
train_padded=pad_sequences(
       train_sequences,maxlen=max_length,padding="post",truncating="post"
)
```

```python
test_sequences=tokenizer.texts_to_sequences(test_sentences)
test_sequences 
```

```python
test_padded=pad_sequences(
       test_sequences,maxlen=max_length,padding="post",truncating="post"
)
```

```python
from keras.models import Sequential
from keras.layers import Embedding,LSTM,Dense,Dropout
model=Sequential()
model.add(Embedding(num_words,32,input_length=max_length))
model.add(LSTM(64)) #lstm remebers the words for a long time
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(train_padded,train_labels,epochs=20,validation_data=(test_padded,test_labels))
```

```python
y_pred=model.predict(test_padded)
y_pred
```
