```python
vocab=nltk.FreqDist(text1) #no.of times each word is occuring #frequency distribution
vocab
```

```python
print(vocab.most_common(20)) #most common word in text1
```

```python
mc=sorted([i for i in vocab.most_common(80) if len(i[0])>3],key=lambda x:x[1],reverse=True) #x[1] have count
mc
```
