```python
import pandas as pd
```

```python
data=pd.read_csv("C:\\Users\\user\\Downloads\\hourly_wages.csv")
data
```

```python
from keras.models import Sequential #2 type model:sequential and functional API
from keras.layers import Dense
#sequential adding layers in stacking up
```

```python
model=Sequential()
```

```python
data.shape
```

```python
#output of first layer is input of second layer
```

```python
X=data.drop(columns=['wage_per_hour'])
X
```

```python
y=data[['wage_per_hour']]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
n_cols=X_train.shape[1]
```

```python
#creating model with layers
model.add(Dense(10,activation="relu",input_shape=(n_cols,))) #relu is used for non linear relationships #input layer
model.add(Dense(10,activation="relu")) #10 is number of nodes in layer
model.add(Dense(1)) #only 1 node in prediction
```

```python
#congigure model
model.compile(optimizer="adam",loss="mean_squared_error") #adam is a commonly used optimizer,learning rate is higher in adam
```

```python
#training the model
from keras.callbacks import EarlyStopping
early_stop=EarlyStopping(patience=3) #patience means after 3 epoch,if there is no improvement in training it will stop there
model.fit(X_train,y_train,epochs=100,callbacks=[early_stop]) #epochs=no.of iteration
```

```python
model.predict(X_test)
```
