```python
import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
x=np.linspace(0,50,50)
y=np.linspace(0,50,50) #linear
```

```python
x+=np.random.uniform(-4,4,50) #adding some noise to non -linear
y+=np.random.uniform(-4,4,50) 
```

```python
n=len(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
X=tf.placeholder("float")
Y=tf.placeholder("float")
```

```python
np.random.randn(2)
```

```python
W=tf.Variable(np.random.randn(),name="W")
b=W=tf.Variable(np.random.randn(),name="b")
```

```python
learning_rate=0.01 #any value in range from 0.001 to 1
```

```python
training_epochs=1000 #how many times your model get trained
```

```python
y_pred=tf.add(tf.multiply(X,W),b)
cost=tf.reduce_sum(tf.pow(y_pred-Y,2))/(2*n)
```

```python
optimizer=tf.train.GradientDescentOptimizer(learning_rate).minimize(cost) #training happens
```

```python
#minimize cost,accuracy increase
```

```python
init=tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    for i in range(training_epochs):
        for (k,l) in zip(x,y):
            sess.run(optimizer,feed_dict={X:k,Y:l})
        c=sess.run(cost,feed_dict={X:x,Y:y})
        print("cost=",c,"W=",sess.run(W),"b=",sess.run(b))
```
