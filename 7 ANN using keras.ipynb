```python
def print_unique_col_values(d):
    for col in d:
        if d[col].dtype=='object':
            print(f'{col}:{d[col].unique()}') #we want to see unique categories for each object column
```

```python
print_unique_col_values(df)
```

```python
df.replace('No internet service','No',inplace=True) #replacing no internet service with no beacause both are same
df.replace('No phone service','No',inplace=True) #replacing no phone service with no beacause both are same
```

```python
print_unique_col_values(df)
```

```python
yes_no_columns=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']
for i in yes_no_columns:
    df[i].replace({'Yes':1,'No':0},inplace=True) #changing yes/no columns to 0 and 1
```

```python
for col in df:
            print(f'{col}:{df[col].unique()}')
```

```python
df['gender'].replace({'Female':1,'Male':0},inplace=True)
df.gender.unique()
```

```python
df2=pd.get_dummies(data=df,columns=['Contract','PaymentMethod','InternetService'])
```

```python
df2.columns
```

```python
df2.dtypes
```

```python
df2['Churn'].replace({"Yes":1,"No":0},inplace=True)
```

```python
df2.dtypes
```

```python
df2.MonthlyCharges
```

```python
cols_to_scale=['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df2[cols_to_scale]=scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col}:{df2[col].unique()}')
```

```python
X=df2.drop('Churn',axis='columns')
y=df2['Churn']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
keras.layers.Dense(26,input_shape=(26,),activation='relu'),
keras.layers.Dense(15,activation='relu'),
keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=30)
```

```python
#confusion matrix in another method
```

```python
yp=model.predict(X_test)
yp
```

```python
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred=[] #actual ytest is range in 0 to 1 ,but we are gettingg in range of 0.3
for i in yp:
    if i>0.3:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
import seaborn as sns
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
